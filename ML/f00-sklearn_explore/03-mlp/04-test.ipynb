# MLP Test 4

Explorando 30 rodadas de Classificação com **MLP** utilizando hiperparâmetros de outros testes.

## Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt

from random import randint
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, recall_score, plot_confusion_matrix, confusion_matrix
```

## Carregando dados

```python
irisDS = load_iris()
X, y = load_iris(return_X_y=True)

acc_list = []
rec_list = []
mat_list = []
seed_list = []
```

## 30 Rodadas de Classificação

```python
for turn in range(30):
    # Gerando semente para distribuição dos dados
    seed = randint(0, 9000)

    while seed in seed_list:
        seed = randint(0, 9000)

    # Distribuindo os dados
    X_pre_train, X_pre_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)

    # Instanciando Standard Scaler
    sc = StandardScaler().fit(X_pre_train)

    # Pré-processando dados
    X_train = sc.transform(X_pre_train)
    X_test = sc.transform(X_pre_test)

    # Instanciando e treinando classificador
    classifier = MLPClassifier(hidden_layer_sizes=(120), activation='tanh').fit(X_train, y_train)

    # Realizando predição
    y_pred = classifier.predict(X_test)

    # Gerando resultados
    acuracia = accuracy_score(y_test, y_pred)
    sensibilidade = recall_score(y_test, y_pred, average=None)
    matrix = confusion_matrix(y_test, y_pred, normalize='true')

    # Adicionando resultados às respectivas listas
    seed_list.append(seed)
    acc_list.append(acuracia)
    rec_list.append(sensibilidade)
    mat_list.append(matrix)

    # Exibindo resultados
    print(f'Rodada: {turn} - seed: {seed} - acurácia: {acuracia}\n')

    print(f'Sensibilidade: {sensibilidade}\n')

    print(classification_report(y_test, y_pred, target_names=irisDS.target_names))
    
    plot_confusion_matrix(classifier, X_test, y_test, normalize='true', display_labels=irisDS.target_names)
    plt.show()
```

```python
metricsDF = pd.DataFrame(data=rec_list, columns=irisDS.target_names)
metricsDF['Acurácia'] = acc_list
metricsDF
```
