# SVM Test

Após o GridSearch, observei uma pequena diferença e ótimos resultados com o kernel `'rbf'` e o `'linear'`, então explorarei 15 rodadas com o `LinearSVC`

## Importando bibliotecas

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from sklearn.svm import SVC, LinearSVC
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, plot_confusion_matrix, recall_score, accuracy_score
```

## Carregando dados

```python
irisDS = load_iris()

X, y = load_iris(return_X_y=True)

seeds = np.random.randint(0, 9000, 15)

print(f'seeds: {seeds}')
```

## 15 rodadas com SVC no kernel `rbf`

```python
rbf_acc, rbf_rec = [], []

for turn, seed in enumerate(seeds):
    # Separando dados
    X_pre_train, X_pre_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)

    # Pré-processando dados
    sc = StandardScaler().fit(X_pre_train)

    X_train = sc.transform(X_pre_train)
    X_test = sc.transform(X_pre_test)

    # Instanciando classificador
    classifier = SVC(
        kernel='rbf',
        decision_function_shape='ovo'
    ).fit(X_train, y_train)

    # Gerando predições e métricas
    y_pred = classifier.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred, average=None)
    setosa, versicolor, virginica = rec

    rbf_acc.append(acc)
    rbf_rec.append(rec)

    # Plotando resultados
    plot_confusion_matrix(
        estimator=classifier,
        X=X_test,
        y_true=y_test,
        display_labels=irisDS.target_names,
        normalize='true'
    )

    print(f'Rodada {turn} - Semente {seed} - Acurácia {acc}')
    print(f'S. Setosa {setosa} - S. Versicolor {versicolor} - S. Virginica {virginica}\n')
    print(classification_report(y_test, y_pred, target_names=irisDS.target_names))
    plt.show()

rbfDF = pd.DataFrame(data=rbf_rec, columns=[f'S - {target}' for target in irisDS.target_names])
rbfDF['Acurácia'] = rbf_acc
```

## 15 rodadas com LinearSVC
