# SVM Test

Experimentando classificação com `SVC`

## Importando bibliotecas

```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, plot_confusion_matrix
```

## Carregando e preprocessando dados

```python
# Carregando dados
irisDS = load_iris()

X, y = irisDS.data, irisDS.target

# Distribuindo dados para treino e teste
X_pre_train, X_pre_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=6)

# Preprocessando dados
sc = StandardScaler().fit(X_pre_train)

X_train = sc.transform(X_pre_train)
X_test = sc.transform(X_pre_test)
```

## Realizando classificação

```python
# Instanciando modelo
classifier = SVC(decision_function_shape='ovo').fit(X_train, y_train)

# Realizando predição
y_pred = classifier.predict(X_test)

# Plotando resultados
plot_confusion_matrix(
    estimator=classifier,
    X=X_test,
    y_true=y_test,
    display_labels=irisDS.target_names
)

print('\n' + classification_report(y_test, y_pred, target_names=irisDS.target_names))
```
