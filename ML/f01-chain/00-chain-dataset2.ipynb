# Chain dataset

Explorando medidas como **kurtosis** e **skewness** no dataset das correntes.

## Importando biblioteca e carregando dados

```python
import pandas as pd

# Carregando DataFrame
chainDF = pd.read_csv('./corrente.csv')

targetS = chainDF.pop('classe')

# Carregando Rótulos das características
feature_names = chainDF.columns

# Carregando rótulos das classes
target_names = ['Normal', 'HI-1', 'HI-2', 'HI-3', 'LI-1', 'LI-2', 'LI-3']

# Plotando histograma das classes
targetS.hist()

# Contando ocorrência das classes
targetS.value_counts().sort_index()
```

Como a assimetria é > 0, significa que, levemente, a "cauda" encontra-se à direita, então a maioria dos dados se concentra mais à esquerda

```python
targetS.skew()
```

Como a curtose é < 0, significa então que a distribuição é mais achatada que uma distribuição normal.

(neste caso, uma distribuição normal é centrada com curtose = 0)

```python
targetS.kurtosis()
```
