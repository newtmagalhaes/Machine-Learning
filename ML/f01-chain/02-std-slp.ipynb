# Standard SLP

## Importando bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import Perceptron
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score, recall_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

## Carregando e preparando dados

```python
# Gerando DataFrame
chainDF = pd.read_csv('./corrente.csv')

targetS = chainDF.pop('classe')

# Carregando Rótulos das características
feature_names = chainDF.columns

# Carregando rótulos das classes
target_names = ['Normal', 'HI-1', 'HI-2', 'HI-3', 'LI-1', 'LI-2', 'LI-3']

# Definindo variáveis
X, y = chainDF.to_numpy(), targetS.to_numpy()

# Definindo array de números aleatórios
seed_array = np.array([4054, 6284, 4916, 2297, 8262, 3342, 7440,  541, 1711, 6730, 8764, 6323, 7498,  885,  771])
```

## 15 rodadas de classificação

```python
acc, rec = [], []

for turn, seed in enumerate(seed_array):
    # Distribuindo os dados
    X_pre_train, X_pre_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)

    # Instanciando Standard Scaler
    sc = StandardScaler().fit(X_pre_train)

    # Pré-processando dados
    X_train = sc.transform(X_pre_train)
    X_test = sc.transform(X_pre_test)

    # Instanciando e treinando classificador
    classifier = Perceptron().fit(X_train, y_train)

    # Realizando predição
    y_pred = classifier.predict(X_test)

    # Gerando resultados
    accuracy = accuracy_score(y_test, y_pred)

    # Adicionando resultados às respectivas listas
    acc.append(accuracy)
    rec.append(recall_score(y_test, y_pred, average=None))

    # Exibindo resultados
    print(f'Rodada: {turn} - seed: {seed} - acurácia: {accuracy}\n')

    print(classification_report(y_test, y_pred, target_names=target_names))
    
    plot_confusion_matrix(
        estimator=classifier,
        X=X_test,
        y_true=y_test,
        normalize='true',
        display_labels=target_names
    )
    plt.show()
```

## Observando métricas de Acurácia e Sensibilidade

```python
metricsDF = pd.DataFrame(data=rec, columns=[f'S - {classe}' for classe in target_names])
metricsDF['Acurácia'] = acc
metricsDF
```

```python
metricsDF.boxplot()
metricsDF.describe()
```

```python
rodadasDF = pd.read_csv('15_rodadas.csv')
rodadasDF['SLP_acc'] = acc
rodadasDF.to_csv(path_or_buf='15_rodadas.csv', index=False)
```
