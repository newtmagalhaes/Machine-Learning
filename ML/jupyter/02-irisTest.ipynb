# IRIS Test

Testando o conjunto de dados [IRIS dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set "wiki of iris flower data set") utilizando as bibliotecas `scikit-learn`, `pandas` e `matplotlib`.

Seguindo o tutorial do [video](https://www.youtube.com/watch?v=39HBlzFV9vk).

## Importando bibliotecas

```python
import sklearn as sk
import pandas as pd

from sklearn.datasets import load_iris
```

## Carregando _DataSet_

Usando `scikit-learn`.

```python
iris_dset = load_iris()

print("target_names:\n", list(iris_dset.target_names))

print("feature_names:\n", list(iris_dset.feature_names))
```

- _target names_: indica quais classificações um dado pode ter;

- _feature names_: indica quais as características que um dado possui;

## Gerando _DataFrame_

Usando `pandas`.

```python
# Gerando dataframe
iris_dframe = pd.DataFrame(data=iris_dset.data, columns=iris_dset.feature_names)

# Adicionando coluna "target name" com classificações dos dados
iris_dframe["target name"] = pd.Categorical.from_codes(iris_dset.target, iris_dset.target_names)

print("Data Frame:\n")
iris_dframe.head(10)
```

## Plotando dataframe em gráfico

Usando `matplotlib`

1. "sepal length (cm)" x "sepal width (cm)"

2. "petal length (cm)" x "petal width (cm)"

```python
iris_dframe.plot.scatter("sepal length (cm)", "sepal width (cm)")
```

```python
iris_dframe.plot.scatter("petal length (cm)", "petal width (cm)")
```
