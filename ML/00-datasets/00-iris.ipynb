# Iris Dataset

Explorando vizualizações do conjunto de dados da iris.

## Importando bibliotecas

```python
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_iris

iris_dataset = load_iris()

irisDF = pd.DataFrame(data=iris_dataset.data, columns=iris_dataset.feature_names)
irisDF
```

```python
irisDF.describe()
```

```python
irisDF.boxplot()
```

```python
irisDF['target'] = iris_dataset.target
irisDF.groupby('target').describe().transpose()
```

```python
irisDF.query('target == 0').boxplot()
```

```python
irisDF.query('target == 1').boxplot()
```

```python
irisDF.query('target == 2').boxplot()
```

```python
sns.pairplot(
    data=irisDF,
    hue='target'
)
```
