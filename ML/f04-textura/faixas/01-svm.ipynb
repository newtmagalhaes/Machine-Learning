# Experimentando Classificação com SVM com 4 classes

## Importando Bibliotecas

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, plot_confusion_matrix

TARGET_NAMES = ['nao_faixa',
                'faixa_frente',
                'faixa_diagonal_direita',
                'faixa_diagonal_esquerda']


def acuracia_por_classe(
        y_true: np.ndarray,
        y_pred: np.ndarray,
        rodada: int = 0) -> pd.DataFrame:

    df = pd.DataFrame(data=[y_true, y_pred], index=[
                      'true', 'pred']).transpose()

    score_dict = {}
    for target, sub_df in df.groupby('true'):
        score_dict[f'{TARGET_NAMES[target]}_quantidade'] = len(sub_df)
        score_dict[f'{TARGET_NAMES[target]}_acc'] = accuracy_score(sub_df['true'].values,
                                                                   sub_df['pred'].values)

    return pd.DataFrame(score_dict, index=[rodada])
```

## Carregando dados

```python
# df tem as ultimas 1000 linhas do csv
# Contendo assim 250 amostras de cada classe
df = pd.read_csv('./csvs/lbp_faixas.csv').tail(1000)

y = df.pop('target').values
X = df.values

# Gerando 30 sementes "aleatórias" dentro do intervalo [0, 8000)
seed_arr = np.random.randint(8000, size=30)
```

## Realizando 30 rodadas de classificação

```python
acc_por_rodada = []
acc_por_classe = []
confu_matrices = []

for rodada, seed in enumerate(seed_arr):
  # separando dados
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=seed)

  clf = make_pipeline(StandardScaler(), SVC()).fit(X_train, y_train)
  y_pred = clf.predict(X_test)
  
  # Adicionando resultados à respectiva lista
  acc_por_rodada.append(
      accuracy_score(
          y_true=y_test,
          y_pred=y_pred))

  acc_por_classe.append(
      acuracia_por_classe(
          y_true=y_test,
          y_pred=y_pred,
          rodada=rodada))

  confu_matrices.append(
      confusion_matrix(
          y_pred=y_pred,
          y_true=y_test,
          normalize='true'))

  print(f'Rodada {rodada}\t-\tSeed {seed}')
  print(
    classification_report(
      y_true=y_test,
      y_pred=y_pred,
      target_names=TARGET_NAMES))

  plot_confusion_matrix(
    estimator=clf,
    X=X_test,
    y_true=y_test,
    display_labels=TARGET_NAMES,
    xticks_rotation=30,
    normalize='true')
  
  plt.show()
```

## Exibindo média e desvio padrão das matrizes de confusão

```python
confusion_matrix_mean = np.mean(confu_matrices, axis=0)
confusion_matrix_std = np.std(confu_matrices, axis=0)

sns.heatmap(
    data=confusion_matrix_mean,
    cmap='Blues',
    vmin=0,
    vmax=1,
    annot=True,
    fmt='.4f',
    mask=confusion_matrix_mean == 0,
    xticklabels=TARGET_NAMES,
    yticklabels=TARGET_NAMES
    ).set(
        title='Matriz de confusão média',
        xlabel='Predicted label',
        ylabel='True label')

plt.show()

sns.heatmap(
    data=confusion_matrix_std,
    cmap='Blues',
    annot=True,
    fmt='.4f',
    mask=confusion_matrix_std == 0,
    xticklabels=TARGET_NAMES,
    yticklabels=TARGET_NAMES
    ).set(
        title='Desvio padrão da matriz de confusão',
        xlabel='Predicted label',
        ylabel='True label')

None
```

## Exibindo Acurácia geral e sua distribuição

```python
acc_s = pd.Series(data=acc_por_rodada, name='Acurácia')

print(acc_s.describe())

acc_s.to_frame().boxplot()

acc_s.to_frame().style.hide_index().background_gradient(cmap='Blues')
```

```python
classes_df = pd.concat(acc_por_classe, ignore_index=True)

acc_columns = [col for col in classes_df.columns if 'acc' in col]
qtd_columns = [col for col in classes_df.columns if 'quantidade' in col]

acc_classes_df = classes_df[acc_columns]
qtd_classes_df = classes_df[qtd_columns]
```
