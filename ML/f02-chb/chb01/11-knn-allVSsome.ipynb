# KNN CHB01 _All VS Some_

Comparando o resultado de 20 rodadas de classificação com KNN entre o modelo treinado com todos os 18 canais e o modelo treinado com apenas os 6 no qual obteve melhor desempenho

## importando bibliotecas

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix
```

## Carregando dados

```python
# Carregando chb
chb_df = pd.read_csv('./chb01_com_dwt.csv')
target_s = chb_df.pop('target')

some_channels = ['2', '3', '6', '7', '10', '14']

# Filtrando colunas do chb que são referentes aos canais em some_channels
some_columns = [column for column in chb_df.columns if column.split('-')[-1] in some_channels]

# Para 20 rodadas com todos os canais
X_all = chb_df.values

# Para 20 rodadas com os 6 canais
X_some = chb_df[some_columns].values

# A classificação não muda para ambos os casos
y = target_s.values

# Carregando sementes
seed_array = np.load('./../20_seeds.npy')
print(f'seeds: {seed_array}')
```

## 20 rodadas com todos os canais

```python
acc_list = []
for turn, seed in enumerate(seed_array):
    # Separando dados
    X_train, X_test, y_train, y_test = train_test_split(X_all, y, test_size=0.3, random_state=seed)

    # Instanciando e treinando classificador
    model = KNeighborsClassifier().fit(X_train, y_train)

    # Realizando predição
    y_pred = model.predict(X_test)

    # Salvando acurácia da predição
    acc = model.score(X_test, y_test)
    acc_list.append(acc)

    print(f'Rodada: {turn}\tSeed: {seed}\tScore: {acc}')

    print(classification_report(
        y_pred=y_pred,
        y_true=y_test,
        target_names=['negative', 'positive'],
        digits=4
    ))

    plot_confusion_matrix(
        estimator=model,
        X=X_test,
        y_true=y_test,
        normalize='true',
        display_labels=['negative', 'positive']
    )
    plt.show()

# Formando dataframe de acurácias
acc_df = pd.DataFrame(data=acc_list, columns=['all channels'])
```

## 20 rodadas com apenas 6 canais

```python
acc_list = []
for turn, seed in enumerate(seed_array):
    # Separando dados
    X_train, X_test, y_train, y_test = train_test_split(X_some, y, test_size=0.3, random_state=seed)

    # Instanciando e treinando classificador
    model = KNeighborsClassifier().fit(X_train, y_train)

    # Realizando predição
    y_pred = model.predict(X_test)

    # Salvando acurácia da predição
    acc = model.score(X_test, y_test)
    acc_list.append(acc)

    print(f'Rodada: {turn}\tSeed: {seed}\tScore: {acc}')

    print(classification_report(
        y_pred=y_pred,
        y_true=y_test,
        target_names=['negative', 'positive'],
        digits=4
    ))

    plot_confusion_matrix(
        estimator=model,
        X=X_test,
        y_true=y_test,
        normalize='true',
        display_labels=['negative', 'positive']
    )
    plt.show()

acc_df['some channels'] = acc_list
```
