# Explorando formato dos dados de fluxo

```python
import numpy as np
import pandas as pd
import seaborn as sns
from zipfile import ZipFile

sns.set(style='darkgrid', rc={'figure.figsize':(12, 6)})
```

```python
with ZipFile('./fluxo.zip') as fluxo:
    print('namelist:\n', fluxo.namelist()[:5])

    csv_path_list = [path for path in fluxo.namelist() if '.csv' in path]
    print('pathlist:\n', csv_path_list[:5])

    # FIRST_CSV_PATH = next(name for name in fluxo.namelist() if '.csv' in name)
    # print(FIRST_CSV_PATH)
```

```python
with ZipFile('./fluxo.zip').open(csv_path_list[0]) as csv_file:
    df = pd.read_csv(csv_file, header=19)
    print(df.head())
```

```python
with ZipFile('./fluxo.zip').open(csv_path_list[1]) as csv_file:
    df = pd.read_csv(csv_file, header=19)
    print(df.head())
```

```python
sns.lineplot(data=df, x='TIME', y='CH1')
```

```python
sns.lineplot(data=df.head(1000), x='TIME', y='CH1')
```
