# GS CV Test 1

Explorando _Grid Search_ e _Cross Validation_ para melhoramento de Hiperparâmetros de classificador **Perceptron* para o conjunto de dados iris

## Importando bibliotecas

```python
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import cross_validate, GridSearchCV, KFold, train_test_split
```

```python
# Carregando dataset
X, y = load_iris(return_X_y=True)

# Instanciando Scaler
sc = StandardScaler().fit(X)

# Pré-processando dados
X_train_trans = sc.transform(X)
```

## K-Fold

a função `split` retorna um array de índices, com 5 splits divide o conjunto de treino em 5 _folders_, uma com índices para teste e os demais índices para treino, os dados de treino como parâmetro ajudam a balancear a divisão, caso seja necessário 

```python
for train, test in KFold(n_splits=5).split(X_train_trans, y):
    print(test)
```

```python
cv_result = cross_validate(
    estimator=Perceptron(),
    X=X_train_trans,
    y=y,
    cv=5
)
mean = cv_result['test_score'].mean()
std = cv_result['test_score'].std()
print(f'mean = {mean}; std = {std}')
cv_result
```

## Grid Search Cross Validation

```python
gscv = GridSearchCV(
    estimator=Perceptron(),
    param_grid={
        'max_iter':[1, 10, 30, 50, 100, 500, 1000],
        'eta0':[0.01, 0.1, 0.5, 1, 2],
    },
    cv=5
).fit(X_train_trans, y)

gscv.best_params_, gscv.best_score_
```

```python
gridDF = pd.DataFrame(gscv.cv_results_)
gridDF
```

```python
gridDF[['params', 'mean_test_score', 'std_test_score', 'rank_test_score']]
```

```python
gridDF[['params', 'mean_test_score', 'std_test_score', 'rank_test_score']].sort_values(by='rank_test_score')
```
